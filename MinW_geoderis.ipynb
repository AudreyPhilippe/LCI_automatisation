```python
num_table=np.linspace(34,102,18)
list_page1_table=['Table '+ str(int(val)) for val in num_table]
num_table2=np.linspace(37,105,18)
list_page4_table=['Table '+ str(int(val)) for val in num_table2]
```

```python
i=0
for table in list_page1_table:
    info_sect=pd.read_excel(path1,sheet_name=table,skiprows=1,header=None,skipfooter=43,usecols=[0,3,17,22,25,31])
    index=list(info_sect.iloc[:,0])+list(info_sect.iloc[:,2])+list(info_sect.iloc[:,4])
    val=list(info_sect.iloc[:,1])+list(info_sect.iloc[:,3])+list(info_sect.iloc[:,5])
    info_sect_new=pd.DataFrame(data=val,columns=['value'],index=index)
    volume_pd=pd.read_excel(path1,sheet_name=table,skiprows=44,header=None,usecols=[0,6,17,25])
    index=list(volume_pd.iloc[:,0])+list(volume_pd.iloc[:,2])
    val=list(volume_pd.iloc[:,1])+list(volume_pd.iloc[:,3])
    volume_pd_new=pd.DataFrame(data=val,columns=['value'],index=index)
    sheet4=list_page4_table[i]
    elements_pd=pd.read_excel(path1,sheet_name=sheet4,skiprows=15,header=[0],usecols=[0,2,3,4,6,7,9,10,12,15,16,18,20],index_col='Identifiant Mesure')
    #coord_pd=pd.read_excel(path1,sheet_name=sheet4,skiprows=4,skipfooter=11,header=[0],usecols=[0,13,17],index_col='Identifiant Mesure')
    #valeur_mine=pd.concat([elements_pd,coord_pd],axis=1)
    i+=1
    writer = pd.ExcelWriter('D:/Documents/philippea/OneDrive - BRGM/FutuRaM/MinW/valeurs_mines_midi_pyrenees.xlsx')
    id_depot=info_sect_new.loc['Identifiant dépôt :'].value
    #info_sect_new.to_excel('D:/Documents/philippea/OneDrive - BRGM/FutuRaM/MinW/valeurs_mines.xlsx')
    info_sect_new.to_excel(writer,id_depot, startrow=1)
    volume_pd_new.to_excel(writer,sheet_name=id_depot,startrow=9)
    elements_pd.to_excel(writer,sheet_name=id_depot,startrow=15)
    writer.close()
```
